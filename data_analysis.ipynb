```python
#Converting the datatype to reduce space complexity
df['Age'] = df['Age'].astype(int)
```

```python
df['Num_Bank_Accounts'] = df['Num_Bank_Accounts'].astype(int)
df['Num_Credit_Card'] = df['Num_Credit_Card'].astype(int)
df['Num_of_Loan'] = df['Num_of_Loan'].astype(int)
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].astype(int)
df['Num_Credit_Inquiries'] = df['Num_Credit_Inquiries'].astype(int)
```

```python
df.info()
```

```python
##Age Distribution: Analyze the distribution of customer ages to identify which age groups are most common.

bins = [0, 20, 30, 40, 50, 60, 70, float('inf')]  
labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71+']

# Categorize age into groups
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Count of people in each age group
age_group_counts = df['Age_Group'].value_counts().reset_index()
age_group_counts.columns = ['Age_Group', 'Count']

print("Age Group Counts:")
print(age_group_counts)
```

```python
#Finding the most commomn occupation
occupation_counts = df['Occupation'].value_counts().reset_index()
occupation_counts.columns = ['Occupation', 'Count']
print("Most Common Occupations:")
print(occupation_counts)
```

```python
# 2. Analyze average income by occupation
occupation_income = df.groupby('Occupation')['Annual_Income'].mean().reset_index()
occupation_income.columns = ['Occupation', 'Average Annual Income']
print("\nAverage Income by Occupation:")
print(occupation_income)
```

```python
# Calculate the correlation between Annual Income and Number of Loans
correlation_income_loans = df['Annual_Income'].corr(df['Num_of_Loan'])
print(f"Correlation between Annual Income and Number of Loans: {correlation_income_loans}")
```

```python
# Segmenting income by age groups (e.g., 18-25, 26-35, etc.)
age_groups = pd.cut(df['Age'], bins=[18, 25, 35, 45, 55, 65, 100], labels=['18-25', '26-35', '36-45', '46-55', '56-65', '66+'])

# Adding a new column for the age group
df['Age_Group'] = age_groups

# Calculate the average income by age group
income_by_age_group = df.groupby('Age_Group')['Annual_Income'].mean()

# Display the result
print(f"Average Income by Age Group: \n{income_by_age_group}")
```

```python
# Average income for each credit score category
income_by_credit_score = df.groupby('Credit_Score')['Annual_Income'].mean()
print(f"Average Income by Credit Score: \n{income_by_credit_score}")
```

```python
# Loan count and credit score relationship
loans_by_credit_score = df.groupby('Credit_Score')['Num_of_Loan'].mean()
print(f"Average Number of Loans by Credit Score: \n{loans_by_credit_score}")
```

```python
# Average number of delayed payments per customer by credit score
delayed_payments_by_credit_score = df.groupby('Credit_Score')['Num_of_Delayed_Payment'].mean()
print(f"Average Number of Delayed Payments by Credit Score: \n{delayed_payments_by_credit_score}")
```

```python
# Average outstanding debt by credit score
avg_outstanding_debt_by_credit_score = df.groupby('Credit_Score')['Outstanding_Debt'].mean()
print(f"Average Outstanding Debt by Credit Score: \n{avg_outstanding_debt_by_credit_score}")
```
